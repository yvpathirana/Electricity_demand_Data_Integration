```python
rain_data= pd.read_excel('./Data/input_data/Rainfall.xlsx')
```

```python
rain_data.head()
```

```python
def melt_drop_table_creater(df):
    df=df.melt(id_vars=['Station_ID','Station_Name','Element_Code','Element_Name','year','month'],value_vars=[f"{i}" for i in range(1,32)],
                                    var_name='day')
    df['day']=pd.to_numeric(df['day'], errors='ignore')
    #dropingg data that does not in time period
    df=df.drop(df[(df['year']<2008) |(df['year']>2010)].index)
    df=df.drop(df[(((df['month']==2) &
                (df['day'] >28 ))|
                ((df['month'].isin([4,6,9,11])) &( df['day']==31)))].index)
    df.value=df.value.replace('-9.9M',np.nan)  
    df["Date"] = pd.to_datetime(df[['month', 'day', 'year']])
    return df
```

```python
def pivot_add_columns(df,e_code,e_name):
    df=df.pivot(index=['Date'],columns=['Station_Name'],values='value').reset_index()
#     df['Element_Code']=[e_code for i in range(730)]
#     df['Element_Name']=[e_name for i in range(730)] ,'Element_Code', 'Element_Name'
    df=df.reindex(columns=['Date','ANURADHAPURA', 'BADULLA', 'BANDARAWELA', 'BATTICALOA',
           'COLOMBO', 'GALLE', 'HAMBANTOTA', 'JAFFNA', 'KATUGASTOTA', 'KATUNAYAKA',
           'KURUNEGALA', 'MAHA ILLUPPALLAMA', 'MANNAR', 'MONERAGALA',
           'NUWARA ELIYA', 'POLONNARUWA', 'POTTUVIL', 'PUTTALAM', 'RATMALANA',
           'RATNAPURA', 'TRINCOMALEE', 'VAVUNIYA' ])
    return df 
```

```python
rain_ymd=melt_drop_table_creater(rain_data) 
rain_ymd.value=rain_ymd.value.replace(['Tr'],0.025)   
rain_ymd.value=rain_ymd.value.replace(['.0T'],0.01) 
rain_ymd.value=rain_ymd.value.replace(['.0R'],0.0) 
rain_ymd=pivot_add_columns(rain_ymd,5,'PRECIP')
for i in rain_ymd.columns[1:]:
    rain_ymd[i]=pd.to_numeric(rain_ymd[i], errors='ignore')
rain_ymd.head() 
```
